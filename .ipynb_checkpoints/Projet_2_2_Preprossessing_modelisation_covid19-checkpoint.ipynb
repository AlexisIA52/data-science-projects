```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
pd.set_option('display.max_row', 111)
pd.set_option('display.max_column', 111)
```

```python
data = pd.read_excel(r"../$ Data_Base/dataset.xlsx")
```

```python
df = data.copy()
```

# Pre-processing

```python
missing_rate = df.isna().sum()/df.shape[0]
```

```python
blood_columns = list(df.columns[(missing_rate < 0.9)  & (missing_rate > 0.88)])
viral_columns = list(df.columns[(missing_rate < 0.80) & (missing_rate > 0.75)])
```

```python
key_columns = ['Patient age quantile', 'SARS-Cov-2 exam result']
```

```python
df = df[key_columns + blood_columns + viral_columns]
df.head()
```

```python
df.shape
```

```python
df['Parainfluenza 2'].value_counts()
```

# TrainTest - Nettoyage - Encodage

```python
from sklearn.model_selection import train_test_split
```

```python
trainset, testset = train_test_split(df, test_size=0.2, random_state=0)
```

```python
trainset['SARS-Cov-2 exam result'].value_counts()
```

```python
testset['SARS-Cov-2 exam result'].value_counts()
```

# Encodage

```python
def encodage(df):
    code = {
        'negative':0,
        'positive':1,
        'not_detected':0,
        'detected':1
    }
    for col in df.select_dtypes('object').columns:
        df[col] = df[col].map(code)

    return df
```

```python
def feature_engineering(df):
    df['est malade'] = df[viral_columns].sum(axis=1) >= 1
    df = df.drop(viral_columns, axis=1)
    return df
```

```python
def imputation(df):
    #df['is na'] = (df['Parainfluenza 3'].isna()) | (df['Leukocytes'].isna())
    #df.fillna(-999)
    df = df.dropna(axis=0)
    return df
```

```python
def preprocessing(df):
    df = encodage(df)
    df = feature_engineering(df)
    df = imputation(df)

    X = df.drop('SARS-Cov-2 exam result', axis=1)
    y = df['SARS-Cov-2 exam result']
    
    print(y.value_counts())
    return X, y
```

```python
X_train, y_train = preprocessing(trainset)
```

```python
X_test, y_test = preprocessing(testset)
```

# Modelisation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.decomposition import PCA
```

```python
model = make_pipeline(PolynomialFeatures(2), SelectKBest(f_classif, k=10),
                      RandomForestClassifier(random_state=0))
```

```python
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from sklearn.model_selection import learning_curve
```

```python
def evaluation(model):
    model.fit(X_train, y_train)
    ypred = model.predict(X_test)

    print(confusion_matrix(y_test, ypred))
    print(classification_report(y_test, ypred))

    N, train_score, val_score = learning_curve(model, X_train, y_train,
                                               cv=4, scoring='f1',
                                               train_sizes=np.linspace(0.1, 1, 10))
    
    plt.figure(figsize=(12, 8))
    plt.plot(N, train_score.mean(axis=1), label='train_score')
    plt.plot(N, val_score.mean(axis=1), label='validation_score')
    plt.legend()                                                
    
```

```python
evaluation(model)
```

### Modélisating

```python
preprocessor = make_pipeline(PolynomialFeatures(2, include_bias=False), SelectKBest(f_classif, k=10))
```

```python
RandomForest = make_pipeline(preprocessor, RandomForestClassifier(random_state=0))
AdaBoost = make_pipeline(preprocessor, AdaBoostClassifier(random_state=0))
SVM = make_pipeline(preprocessor, StandardScaler(), SVC(random_state=0))
KNN = make_pipeline(preprocessor, StandardScaler(), KNeighborsClassifier())
```

```python
list_of_models = {'RandomForest': RandomForest,
                  'AdaBoost': AdaBoost,
                  'SVM': SVM,
                  'KNN': KNN}
```

```python
for model in list_of_models: 
    print(model)
    evaluation(list_of_models[model])
```

### Optimisation

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
hyper_params = {'svc__gamma': [1e-2, 1e-3, 1e-4],
               'svc__C': [1, 10, 100, 1000]}
```

```python
grid = GridSearchCV(SVM, hyper_params, scoring='recall', cv=4)

grid.fit(X_train, y_train)

print(grid.best_params_)

y_pred = grid.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
evaluation(grid.best_estimator_)
```

```python
hyper_params = {'svc__gamma': [1e-2, 1e-3, 1e-4],
               'svc__C': [1, 10, 100, 1000],
               'pipeline__polynomialfeatures__degree': [2, 3, 4],
               'pipeline__selectkbest__k': range(40, 60)}
```

```python
grid = RandomizedSearchCV(SVM, hyper_params, scoring='recall', 
                          cv=4, n_iter=40)

grid.fit(X_train, y_train)

print(grid.best_params_)

y_pred = grid.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
evaluation(grid.best_estimator_)
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
precision, recall, threshold = precision_recall_curve(y_test, grid.best_estimator_.decision_function(X_test))
```

```python
plt.plot(threshold, precision[:-1], label='precision')
plt.plot(threshold, recall[:-1], label= 'recall')
plt.legend()
```

```python
def model_final(model, X, threshold=0):
    return model.decision_function(X) > threshold
```

```python
y_pred = model_final(grid.best_estimator_, X_test, threshold=-1)
```

```python
f1_score(y_test, y_pred)
```
